```python
import numpy as np
import matplotlib.pyplot as plt

# define a function to draw double-headed arrows (for coordinate axes)
# **kwargs allow you to include keyword arguments like head_width, head_length, color, etc. and pass them to plt.arrow
def double_headed_arrow(x1, y1, x2, y2, **kwargs): 
    plt.arrow(x1, y1, x2 - x1, y2 - y1, **kwargs)
    plt.arrow(x2, y2, x1 - x2, y1 - y2, **kwargs)

# create a set of points to transform
v = np.array([[0, 3],
              [0, 4]])

# define a transformation matrix to reflect v over the line y=x
A = np.array([[0, 1],
              [1, 0]])

# apply the transformation
T_v = A @ v

# plot the original and transformed points
plt.figure()

plt.title("Mr. Shadle's linear transformation for #6\nLin. Alg. for Data Science\nFall 2024")

plt.arrow(0, 0, v[0][1], v[1][1], head_width=0.15, width=0.05, color='b', label='$\\vec v$')
plt.arrow(0, 0, T_v[0][1], T_v[1][1], head_width=0.15, width=0.05, color='r', label='$T(\\vec v)$')

# draw the line y=x as a dotted line since this problem involves reflection over the line y=x
plt.plot([-1, 5], [-1, 5], 'k:', label=f'$y=x$')

# annotate the coordinates of the arrowheads
plt.text(v[0][1], v[1][1], f'   ({v[0][1]}, {v[1][1]})',
         horizontalalignment='left',
         verticalalignment='bottom')
plt.text(T_v[0][1], T_v[1][1], f'   ({T_v[0][1]}, {T_v[1][1]})',
         horizontalalignment='left',
         verticalalignment='bottom')

# annotate the x and y axes
plt.text(5.2, 0, 'x',
         horizontalalignment='left',
         verticalalignment='bottom')
plt.text(0, 5.2, 'y',
         horizontalalignment='left',
         verticalalignment='bottom')

# The "old" way of drawing coordinate axes
# plt.vlines(0.0, -5, 5, linestyle='-', color = 'k') # '-', '--', '-.', ':', '',
# plt.hlines(0.0, -5, 5, linestyle='-', color = 'k')

# The new and improved way of drawing coordinate axes with arrowheads
double_headed_arrow(-1, 0, 5, 0, head_width=0.1, head_length=0.2, color='black')
double_headed_arrow(0, -1, 0, 5, head_width=0.1, head_length=0.2, color='black')

# Make sure the plot is "square" rather than skewed
plt.axis('square')

# Include a legend
plt.legend()

# Include a grid
plt.grid()

# Show the plot
plt.show()
```
